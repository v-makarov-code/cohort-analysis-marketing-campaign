```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors
import numpy as np
import pandas as pd
import pandahouse as ph
```

```python
#параметры соединения - нужны, чтобы подключиться к нужной схеме данных
connection = {'host': 'https://clickhouse.lab.karpov.courses',
'database':'simulator_20250520',
'user':'student',
'password':'dpo_python_2020'
}

#текст запроса
query = '''
SELECT start_day, day, source, count(user_id) AS users
  FROM 
    (SELECT user_id, min(toDate(time)) AS start_day
       FROM simulator_20250520.feed_actions
      GROUP BY user_id) t1
      
       JOIN
       
    (SELECT DISTINCT user_id, toDate(time) AS day, source 
       FROM simulator_20250520.feed_actions) t2
      USING user_id 
      WHERE source = 'ads'
         
 GROUP BY start_day, day, source
'''

# эта функция выполнит запрос и запишет его результат в pandas DataFrame
df = ph.read_clickhouse(query, connection=connection)
```

```python
df
```

### Когортный анализ

#### Платный трафик

```python
df.dtypes
```

```python
df = df.assign(cohort_period = (df.day - df.start_day).dt.days)
```

```python
df.head()
```

```python
cohort_pivot = df.pivot_table(index='start_day', columns='cohort_period', values='users')
```

```python
cohort_pivot.head(10)
```

```python
# Рассчитываем размеры когорт (первый столбец сводной таблицы)
cohort_size = cohort_pivot.iloc[:, 0]
```

```python
# Вычисляем коэффициенты удержания, деля на размер когорты
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
```

```python
retention_matrix
```

```python
retention_matrix.iloc[29, 20] # 20 день 30 когорты
```

#### Какие пользователи не смогли воспользоваться лентой 2025-05-24

Из графика DAU видно, что падение в активных пользователях на 24 мая было больше 2 тысячи человек относительно предыдущего дня.
Такое падение возможно только из проблем у пользователей из России, поэтому фильтр в запросе будет стоять по ним. 
Дальше детализируем, какие именно пользователи из России имели проблемы в этот день.

```python
#текст запроса
query = '''
SELECT country, city, os, COUNT(DISTINCT user_id)
  FROM simulator_20250520.feed_actions 
 WHERE toStartOfDay(toDateTime(time)) = toDateTime('2025-05-23')
   AND country = 'Russia'
 GROUP BY country, city, os
 ORDER BY COUNT(DISTINCT user_id) DESC
'''

# эта функция выполнит запрос и запишет его результат в pandas DataFrame
df_20250523 = ph.read_clickhouse(query, connection=connection)
```

Сраним срез пользователей на 23 мая с пользователями на 24

```python
df_20250523.head() # Срез пользователей из России на 23 мая
```

```python
df1 = df_20250523.groupby("city").agg(
    sum_users = ("uniqExact(user_id)", "sum")).reset_index().sort_values(
    by="sum_users", ascending=False).head(10)
```

```python
df1.head() # Суммарное кол-во пользователей для каждого города
```

```python
#текст запроса
query = '''
SELECT country, city, os, COUNT(DISTINCT user_id)
  FROM simulator_20250520.feed_actions 
 WHERE toStartOfDay(toDateTime(time)) = toDateTime('2025-05-24')
   AND country = 'Russia'
 GROUP BY country, city, os
 ORDER BY COUNT(DISTINCT user_id) DESC
'''

# эта функция выполнит запрос и запишет его результат в pandas DataFrame
df_20250524 = ph.read_clickhouse(query, connection=connection)
```

```python
df_20250524.head()  # Срез пользователей из России на 24 мая
```

```python
df2 = df_20250524.groupby("city").agg(
    sum_users = ("uniqExact(user_id)", "sum")).reset_index().sort_values(
    by="sum_users", ascending=False).head(10)
```
